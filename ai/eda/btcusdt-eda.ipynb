```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
btc_usdt_df = pd.read_csv("../data/BTCUSDT-1m-data.csv", index_col=0)
display(btc_usdt_df.head(10))
display(max(btc_usdt_df["timestamp"]))
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=btc_usdt_df, x="timestamp", y="close", hue="volume", ax=ax)
sns.set_xticks()
plt.xlabel("Timestamp")
plt.ylabel("Close ($)")
plt.show();
```
